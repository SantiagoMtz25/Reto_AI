###Description of the data

Now that we can see that:

- Numerical: PassengerId, Survived, Pclass, Age, Sibsp, Parch, Fare
    - Continous: Age, PassengerId, SibSp, Parch, Fare
    - Discrete: Survived, Pclass
- Categorical: Name, Sex, Ticket, Cabin, Embarked

What does it mean each piece of code:

- Passenger ID: A unique identifier for each passenger
- Survived: Whether the passenger survived (1 = Survived, 0 = Did not survive)
- Pclass: The passenger's class on the ship (1 = 1st class, 2 = 2nd class, 3 = 3rd class). -> It represents socio-economic status
- Name: The name of the passenger
- Sex: The gender of the passenger (Male/Female)
- Age: The age of the passenger in years.
- SibSp: The number of siblings or spouses the passenger had aboard
- Parch: The number of parents or children the passenger had aboard
- Ticket: The ticket number of the passenger (dropped in your version)
- Fare: The amount of money the passenger paid for the ticket
- Cabin: The cabin number the passenger stayed in
- Embarked: The port where the passenger boarded the Titanic (C = Cherbourg, Q = Queenstown, S = Southampton).

## Random Forest

Justification: Given the Titanic datasets diversity in features (like age, fare, class, etc.), Random Forest can effectively capture the non-linear relationships between features and survival.

```python
# Define the parameter grid for Random Forest
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees in the forest
    'max_depth': [5, 10, 15],  # Maximum depth of each tree
    'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4], # Minimum samples required for node
    'max_features': ['auto', 'sqrt', 'log2'], # Features considered at each split
}

# Create a Random Forest Classifier object
rf = RandomForestClassifier()

# Perform Grid Search with cross-validation
grid_search_rf = GridSearchCV(rf, param_grid, cv=5, scoring='accuracy')  # Adjust 'scoring' as needed
grid_search_rf.fit(df_pca_train, y_train)
```

```python
# Chosen hyperparameters
rf = RandomForestClassifier(n_estimators=100, max_depth=5, min_samples_split=10,
                           min_samples_leaf=2, max_features='log2')

rf.fit(df_pca_train, y_train)
```

```python
results = pd.DataFrame(grid_search_rf.cv_results_)
hyperparameters = ['n_estimators', 'max_depth', 'min_samples_split', 'min_samples_leaf', 'max_features']
# Create a DataFrame to store the results for each hyperparameter
hyperparameters_results = pd.DataFrame(columns=hyperparameters)

# Iterate over each combination of hyperparameters and store the rank
for i in range(len(results)):
    for param, value in results.iloc[i]['params'].items():
        if param in hyperparameters:
            hyperparameters_results.loc[i, param] = value
    hyperparameters_results.loc[i, 'rank'] = results.iloc[i]['rank_test_score']

hyperparameters_results

# Plot each hyperparameter against the rank
for param in hyperparameters:
    plt.figure()
    plt.scatter(hyperparameters_results[param], hyperparameters_results['rank'])
    plt.xlabel(param)
    plt.ylabel('Rank')
    plt.title(f'Relationship between {param} and rank')
    plt.show()
```

1. Gamma (gamma)

Influencia de los puntos de entrenamiento: gamma define qu tan lejos llega la influencia de un solo punto de entrenamiento.
Un valor bajo de gamma significa una influencia ms amplia, lo que resulta en un modelo ms suave que captura patrones ms generales en los datos.
Un valor alto de gamma significa una influencia ms localizada, lo que resulta en un modelo ms complejo que se ajusta ms estrechamente a los puntos de entrenamiento individuales.
Relacin con el sobreajuste:
Un gamma demasiado alto puede llevar a un sobreajuste, donde el modelo se adapta demasiado bien a los datos de entrenamiento y tiene un rendimiento deficiente en datos nuevos.
Un gamma demasiado bajo puede llevar a un subajuste, donde el modelo es demasiado simple para capturar la complejidad de los datos.

2. Kernel (kernel)

Funcin de transformacin: El kernel define la funcin matemtica utilizada para transformar los datos de entrada en un espacio de mayor dimensin, donde se puede encontrar un hiperplano lineal que separa las clases.
Kernel RBF ('rbf'): El kernel de funcin de base radial (RBF) es uno de los kernels ms populares en SVM. Crea un espacio de caractersticas de dimensin infinita, lo que permite al modelo aprender relaciones no lineales complejas en los datos.

3. C (C)

Regularizacin: C controla el equilibrio entre maximizar el margen entre las clases y minimizar el error de clasificacin en los datos de entrenamiento.
Relacin con el sobreajuste y el subajuste:
Un valor alto de C penaliza fuertemente los errores de clasificacin, lo que puede llevar a un margen ms estrecho y potencialmente a un sobreajuste.
Un valor bajo de C permite un margen ms amplio pero puede llevar a una mayor clasificacin errnea de los puntos de entrenamiento y potencialmente a un subajuste.
